1. Data cleanup
Data was exported from the department of labor website. I cleaned up data in MATLAB and exported to a .csv file. There are about ~40 fields in the original csv however I keep only the ones tha I thought would be interesting. In the final file I have:
    1. Case number
    2. Certified/denied - application status
    3. Employer name
    4. Employer city
    5. Employer state
    6. Employer zip
    7. Employment start date
    8. Employment end date
    9. File date
    10. Decision date
    11. Job title
    12. Job SOC
    13. Wages
    14. Full time status
    15. Rate of pay
    16. Number of workers
